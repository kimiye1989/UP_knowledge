```python
import pandas as pd
from textblob import TextBlob
import re
from collections import Counter
import difflib
```

```python
def clear(word):
    for ch in '!"#$&()*+,，-./:;<=>?@[\\]^_{|}·~‘’':
        word=word.replace(ch," ")
    word=re.sub('[\u4e00-\u9fa5]', '', word)
    # word=word.lower()
    word=word.replace("pdf","")
    word=word.replace("  "," ")
    word=word.replace("  "," ")
    word=word.replace("  "," ")
    word=TextBlob(word).correct()
    word=str(word)
    # print(type(x))
    return word
```

```python
# title_all=TextBlob(title_str)
# title_all.noun_phrases
y=1923
y_twords={}

x=[]
for i in range(df.shape[0]):

    year=df['年份'][i]
    title=df['文章题目'][i]
    content=clear(title)
    if year ==y:
        x.append(content)
        x_str=" ".join(x)
        y_twords[y]=TextBlob(x_str).noun_phrases
    else:
        y=y+1
        x=[]
        x.append(content)
# print(y_words)
y_tcount={}
for key,value in y_twords.items():
    y_tcount[key]=Counter(value)
```

```python
from simhash import Simhash
def simhash_demo(text1, text2):
    """
    求两文本的相似度
    :param text1:
    :param text2:
    :return:
    """
    a_simhash = Simhash(text1)
    b_simhash = Simhash(text2)
    max_hashbit = max(len(bin(a_simhash.value)), (len(bin(b_simhash.value))))
    # 汉明距离
    distince = a_simhash.distance(b_simhash)
    # print(distince)
    similar = 1 - distince / max_hashbit
    return similar
# simhash_demo('city planning','planning city')
def replace_similar(lists,similarity=0.9):
    i = 0
    # j = 1
    while i<len(lists):
        for j in range(i+1,len(lists)):
            ratio=simhash_demo(lists[i],lists[j])
            if ratio>similarity:
                lists[j]=lists[i]
                # print(j)
        i+=1
    return lists

# print(len(y_words[1974]))
```
